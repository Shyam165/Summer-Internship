```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('classic')
# style.use('dark_background')

x=[2,4,6,8]
y=[10,23,43,33]

x1=[4,8,24]
y1=[3,6,9]

x2=[6,12,18]
y2=[12,24,36]

plt.bar(x,y)
plt.bar(x1,y1)
plt.bar(x2,y2)

plt.title('test')
plt.xlabel('test for x')
plt.ylabel('test for y')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('classic')

number=[1,2,3,4,5,3,2,3,4,5,4,2,4,4,33,22,4,5,3,65,43,34,22,111,66,76,55,50,40,30,70]
jumps=[0,15,30,45,60,75,90,105]

plt.title('test 1 hist')
plt.xlabel('test for x')
plt.ylabel('test for y')

plt.hist(number,jumps,histtype='step') #bar
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('classic')

food=['pizza', 'ice cream', 'burgers']
sales=[10,25,40]
color=['red', 'green', 'yellow']

plt.pie(sales,labels=food, colors=color)

plt.title('example food')
plt.show()
```
